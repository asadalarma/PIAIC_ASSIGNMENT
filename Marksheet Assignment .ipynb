```python

#Marksheet 
import csv 

print('\n==================================PROGRAM USING LISTS ARRAY======================\n') 

number_of_subjects=int(input('Please enter number of subjects in digits form: '))
subjects_array=[]
subjects_marks_array=[]
grades_array=[]
one_subject_total_marks = 100
total_marks_obtained = 0

for subjects_i in range(number_of_subjects):
    name_of_subjects=str(input('Please enter name of the subject %d: ' %(subjects_i+1)))
    subjects_array.append(name_of_subjects)

print('\n')

for subjects in subjects_array:
    marks_of_subjects=float(input('Please enter marks for %s: '%(subjects)))
    subjects_marks_array.append(marks_of_subjects)

    if marks_of_subjects >=40 and  marks_of_subjects <=49:
        grades_array.append('D')
        
    if marks_of_subjects >=50 and  marks_of_subjects <=59:
        grades_array.append('C')
        
    elif marks_of_subjects >=60 and  marks_of_subjects <=69:
        grades_array.append('B')
    
    elif marks_of_subjects >=70 and  marks_of_subjects <=79:
        grades_array.append('A')
    
    elif marks_of_subjects >=80 and  marks_of_subjects <=89:
        grades_array.append('A-')
        
    elif marks_of_subjects >=90 and  marks_of_subjects <=100:
        grades_array.append('A+')

print('\n==================================THE SUBJECTS ARE==============================\n')    
print(subjects_array)
print('\n==================================THE MARKS ARE=================================\n')  
print(subjects_marks_array)
print('\n==================================THE GRADES FOR MARKS ARE======================\n')  
print(grades_array)

print('\n==================================THE MARKSHEET=================================\n') 


print("\n\n SUBJECTS \t\tMARKS OBTAINED \t\tTOTAL MARKS \tGRADE")

with open("marksheet.csv", "w", newline="") as f: 
    data_handler = csv.writer(f, delimiter=",")     
    data_handler.writerow(["SUBJECTS", "MARKS OBTAINED", "TOTAL MARKS","GRADE"])

    for i in range(number_of_subjects):
        total_marks_obtained += subjects_marks_array[i]
        print("\n %s \t\t%.2f \t\t%.1f   \t\t%s" %(subjects_array[i],subjects_marks_array[i],one_subject_total_marks,grades_array[i]))
        data_handler.writerow([subjects_array[i], subjects_marks_array[i], one_subject_total_marks,grades_array[i]])

        
    print('\n==================================THE MARKSHEET ENDS=================================\n') 
    
    percentage = (total_marks_obtained / (one_subject_total_marks * number_of_subjects )) * 100
    print('\n================================== YOUR PERCENTAGE IS ===========================\n',percentage,"%")
    data_handler.writerow([])
    data_handler.writerow(["PERCENTAGE", "", "",str(percentage) + str(" %")])

    
```
